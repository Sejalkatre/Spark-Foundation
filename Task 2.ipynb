```python
# We are changing names of columns because the dataset given to us haslots of unusable columns and 
# name also not written in proper condition.

df.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude',
    'longitude':'longitude','summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound'},inplace=True)
```

```python
# we are making new dataset form folling given valriable which is usefull tp us

df = df[['Year','Month','day','Country','State','Region','City','latitude','longitude',"Attacktype",'kill',
               'Wound','target1','summary','Group','Targettype','Weapon','motive','ransompaid','ransomamt']]
```

```python
# We are viewing new dataset after change 
df
```

```python
plt.figure(figsize = (15,5))
sns.countplot(x = 'Region',data = df,palette = 'rocket_r',order = df["Region"].value_counts().index)
plt.xticks(rotation = 90)
plt.show()
```

## Observations :

- Middle East & North Africa, South Asia region has highest no of terror attack
- Australasia & Oceania region has lowest no if terror attack


```python
plt.figure(figsize = (10,12))
sns.countplot(x = "Attacktype",data = df,order = df['Attacktype'].value_counts().index)
plt.xticks(rotation = 90)
plt.show()
```

## Observations :

- Most of the terror attacks are done by Bombarding and Explosion 
- Very less terror attacks are done by Hijacking. Because airports have high security

```python
plt.figure(figsize = (10,12))
sns.countplot(x = 'Year',data = df,palette = 'rocket_r',order = df['Year'].value_counts().index)
plt.title('Attacks per Year',fontsize = 20)
plt.ylabel('No of Attacks')
plt.xticks(rotation = 90)
plt.show()
```

## Observation :

- Most of the terror attacks occurs in year 2014.

```python
plt.figure(figsize = (5,5))
country_kill = df.groupby('Year')['kill'].sum()
top_10_country = country_kill.sort_values(ascending = False).iloc[0:10]
top_10_country.plot(kind = 'barh')
plt.title('People Died Due to Attack')
plt.show()
```

## Observation :

- Most of the people died in the year 2014.

```python
total_ransom = df.groupby('Country')['ransomamt'].sum()
top_10_country = total_ransom.sort_values(ascending = False).iloc[0:10]
top_10_country.plot(kind = 'barh')
plt.title('Top 10 Ransome Given Country')
plt.show()
```

## Observation :

- Italy has given the most ransome as compared to other countries.

```python
plt.subplots(figsize=(20,10))
sns.barplot(x = df['Country'].value_counts()[:10].index, y = df['Country'].value_counts()[:10].values,palette='twilight_r')
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation = 50)
plt.show()
```

## Observation : 

- Most of the terror attacks took in Iraq, Pakistan , Afghanistan and India 

```python
plt.subplots(figsize = (20,10))
sns.barplot(x = df['City'].value_counts()[:10].index, y = df['City'].value_counts()[:10].values,palette = 'cividis')
plt.title('Top Cities Affected')
plt.xlabel('Cities')
plt.ylabel('Count')
plt.xticks(rotation = 50)
plt.show()
```

## Observation :

- Most of the attack occurs in unknown cities

```python
plt.figure(figsize = (20,10))
sns.barplot(x = df['Group'].value_counts()[:12].index, y = df['Group'].value_counts()[:12].values,palette = 'cubehelix')
plt.title('Groups behind all attacks')
plt.xlabel('Groups')
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.show()
```

## Observations :

- Most of the terror attacks are done by unknown terrorist groups.
- Taliban , (ISIL) , (Sl) groups are activated in large numbers.

```python
plt.figure(figsize=(20,10))
sns.barplot(x = df['Targettype'].value_counts()[:10].index,y = df['Targettype'].value_counts()[:10].values,palette = 'rocket_r')
plt.title('Targeted People',fontsize = 20)
plt.xlabel('Targets',fontsize = 15)
plt.ylabel('count',fontsize=15)
plt.xticks(rotation = 90)
plt.show()
```

## Observations :

- Terrorist has targeted the most Private Citizens and Properties
- The second target was Military peoples.  

```python
df[['Group','kill']].groupby(['Group']).sum().drop('Unknown').sort_values('kill',ascending=False).head(10).plot(kind='barh',color='gold',figsize=(20,10),fontsize = 15)
plt.title("Top 10 terrorist group attack",fontsize=20)
plt.xlabel(" No of Killed People",fontsize=15)
plt.ylabel(" Terrorist group name",fontsize=15)
plt.show()
```

# Conclusion :

- Country with the most attacks: Iraq
- City with the most attacks: Baghdad
- Hot Zones for Terrorist : Middle East & North Africa
- Year with the most attacks: 2014
- Group with the most attacks: Unknown & Taliban
- Most Attack Types: Bombing/Explosion
